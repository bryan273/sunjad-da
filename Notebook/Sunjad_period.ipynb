````python
courses = pd.read_csv(HOME+"courses.csv")
classes = pd.read_csv(HOME+"classes.csv")
````

````python
classes.head()
````

````python
df = pd.DataFrame()
new_rows = []

def get_lecturers(row):
  lecturers = ast.literal_eval(row['lecturer'])
  lecturers = [lecturer.strip() for lecturer in lecturers]
  new_lecturers = [ [lecturer, row['course_id'], row['name'], row['period_id'], row['schedule_items']] for lecturer in lecturers]
  for new_lecturer in new_lecturers:
    new_rows.append(new_lecturer)

classes.progress_apply(get_lecturers, axis=1)
df = df.append(new_rows).rename(columns={0: 'lecturer', 1:'course_id', 2:'course_name', 3:'period_id', 4:'schedule_items'})
df.to_csv(HOME+'lecturers.csv', index=False)
df
````
